---

## üì¶ Paso 1: Importar pandas

```python
import pandas as pd
```

---

## üìÅ Paso 2: Cargar el archivo Excel

Aseg√∫rate de haber subido tu archivo (`REPORTE VOLUMEN DE CREDITO 2024_12.xlsx`) a Colab antes de ejecutar esta celda.

```python
df = pd.read_excel('REPORTE VOLUMEN DE CREDITO 2024_12.xlsx', skiprows=8)
df.head()
```

---

## üßæ Paso 3: Ver nombres de columnas

```python
df.columns
```

---

## ‚úçÔ∏è Paso 4: Renombrar columnas

```python
df = df.rename(columns={
    'N√öMERO DE OPERACIONES': 'Num_Operaciones',
    'MONTO OTORGADO': 'Monto_Otorgado'
})
df.columns
```

---

## üßπ Paso 5: Eliminar columnas innecesarias

```python
df = df.drop(columns=['SUBSISTEMA'], errors='ignore')
```

---

## üîÑ Paso 6: Convertir tipos de datos

```python
df['FECHA'] = pd.to_datetime(df['FECHA'], errors='coerce')
df['Monto_Otorgado'] = pd.to_numeric(df['Monto_Otorgado'], errors='coerce')
df['Num_Operaciones'] = pd.to_numeric(df['Num_Operaciones'], errors='coerce')
df.dtypes
```

---

## üï≥ Paso 7: Detectar valores faltantes

```python
df.isnull().sum()
```

---

## üõ† Paso 8: Rellenar o eliminar valores vac√≠os

```python
# Eliminar filas con valores faltantes
# df = df.dropna()

# Rellenar con texto
df['PROVINCIA'] = df['PROVINCIA'].fillna('Sin dato')

# Rellenar con promedio
df['Monto_Otorgado'] = df['Monto_Otorgado'].fillna(df['Monto_Otorgado'].mean())
```

---

## ‚ûï Paso 9: Crear una nueva columna

```python
df['Monto_Promedio'] = df['Monto_Otorgado'] / df['Num_Operaciones']
df[['PROVINCIA', 'Monto_Promedio']].head()
```

---

## ‚úÖ Paso final: Guardar el archivo limpio (opcional)

```python
df.to_csv('datos_limpios.csv', index=False)
```
